```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Create the vectors X and Y
x = np.array(range(100))
y = (10 -x) /3
y2 = (37 -5*x) / 8

# Create the plot
plt.plot(x,y,label='y = 10 -x /3')
plt.plot(x,y2,label='y2 = (37 - 5*x) / 8')

# Add a title
plt.title('My first Plot with Python')

# Add X and y Label
plt.xlabel('x axis')
plt.ylabel('y axis')

# Add a grid
plt.grid(alpha=.4,linestyle='--')

# Add a Legend
plt.legend()

# Show the plot
plt.show()
```
